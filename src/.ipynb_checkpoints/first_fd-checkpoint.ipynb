## FD for the first order derivative

Given a smooth function $f: \mathbb R \mapsto \mathbb R$, its derivative is
$$f'(x) = \lim_{h\to 0} \frac{f(x+h) - f(x)}{h}.$$
This means, if $h$ is small enough, then
$$f'(x) \simeq  \frac{f(x+h) - f(x)}{h} := \delta_h f.$$
We call $\delta_h$ by Finite Difference (FD) operator. In particular, 
- If $h>0$, then $\delta_h$ is Forward Finite Difference (FFD);
- If $h<0$, then $\delta_h$ is Backward Finite Difference (BFD);
- The average of FFD and BFD is Central Finite Difference (CFD), denoted by
$$\delta_{\pm h} f (x) := \frac 1 2 (\delta_h f (x) + \delta_{-h} f(x)) \simeq f'(x).$$


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**ex1** Let $f(x) = \sin x$. Plot $f'$, $\delta_h f$, $\delta_{-h}f$, and $\delta_{\pm}f$ with $h = .5$

```python
def FFD(f, x, h):
    return (f(x+h) - f(x))/h

def BFD(f, x, h):
    return (f(x) - f(x-h))/h

def CFD(f, x, h):
    return (f(x+h) - f(x-h))/h/2

h = .5

x_co = np.linspace(0, 2*np.pi, 100)
plt.plot(x_co, np.cos(x_co), label = 'cosine');
plt.plot(x_co, FFD(np.sin, x_co, h), label = 'FFD');
plt.plot(x_co, BFD(np.sin, x_co, h), label = 'BFD');
plt.plot(x_co, CFD(np.sin, x_co, h), label = 'CFD');
plt.legend();
```

## FD for the second order derivative

One of the commonly used FD for $f''$ is the following:
$$f''(x) = \frac{d}{dx} f'(x) \simeq \delta_h f'(x) \simeq \delta_h \delta_{-h} f(x) := \delta_h^2 f(x).$$
If we write it explicitly, then
$$f''(x) \simeq \delta_h^2 f(x) = \frac{f(x+h) - 2 f(x) + f(x-h)}{h^2}.$$

**ex2** Let $f(x) = \sin x$. Plot $f''$ and $\delta_h^2 f$ with $h = .5$

```python
def SFD(f, x, h):
    return (f(x+h) - 2*f(x) + f(x-h))/np.power(h,2)

h = .5

x_co = np.linspace(0, 2*np.pi, 100)
plt.plot(x_co, - np.sin(x_co), label = '$-sin(x)$');
plt.plot(x_co, SFD(np.sin, x_co, h), label = 'SFD');
plt.legend();
```

## Partial differentiation with FD

Given a smooth function $f(t, x): \mathbb R^2 \mapsto \mathbb R$, we write
$$\partial_t f(t, x) \simeq \delta_{t, \tau} f(t, x) = \delta_\tau f(\cdot, x)(t) = \frac{f(t+\tau, x) - f(t, x)}{\tau}.$$
Similarly, we have
$$\partial_x f(t, x) \simeq \delta_{x, h} f(t, x) = \delta_h f(t, \cdot) (x),$$
and
$$\partial_{x}^2 f(t,x) \simeq \delta_{x, h}^2 f(t, x) = \delta_h^2 f(t, \cdot) (x).$$
