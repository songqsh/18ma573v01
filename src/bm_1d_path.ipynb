# Generate BM path

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
'''=========
generate 1d bm path
========='''

def bm_1d_path(time_grid):
    step_size = np.diff(time_grid)
    dw = np.random.normal(0, np.sqrt(step_size)) #bm increment
    w = np.append([0], np.cumsum(dw)) #bm path
    return w
```

```python
'''===========
test bm_1d_path:
    plot multiple bm path
==========='''
terminal = 1.
num_steps = 10
time_grid = np.linspace(0, terminal, num_steps + 1)

num_path = 5
for i in range(num_path):
    bm1 = bm_1d_path(time_grid) #generate path                               
    plt.plot(time_grid, bm1)
```
