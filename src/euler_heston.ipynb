## General 2-d SDE

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

'''=============
test SDE_2d.euler
plot paths
=============='''
sde1 = Sde_2d(init_state=[100., .04], 
              drift=lambda x:[.01, .03], 
              vol= lambda x:[[.1, 0],[0, .1]]
              )
grid = np.linspace(0,1,100)
num_path = 5
path = [sde1.euler(grid) for i in range(num_path)] #path container


plt.figure()


plt.subplot(2,1,1)
plt.title('Sde_2d.euler')
plt.ylabel('1st state')
for i in range(num_path):
    plt.plot(grid, path[i][0]) #first dimension of the path

plt.subplot(2,1,2)
plt.ylabel('2nd state')
for i in range(num_path):
    plt.plot(grid, path[i][1]) #first dimension of the path
```

## Heston option price by Euler scheme

We implement ordinary MC to find Heston call price, compare with exact price, see [pdf](../doc/euler_sde_2d.pdf)

```python
from contract_v01 import VanillaOption
```

```python
#initiate option and heston instances
heston1 = Heston(init_state=[100., .04], r=.05, 
                 kappa=1.2, theta=.04, xi=.3, rho=.5)
option1 = VanillaOption(otype = 1, strike = 100., maturity= 1., market_price=15.) 
```

```python
'''
==========
option pricing based euler OMC
=========='''

r = heston1.drift([1,1])[0] #rate will be used for discount purpose

num_step = 10
time_grid = np.linspace(0, option1.maturity, num_step+1)

num_path = 10000
terminal_price = [heston1.euler(time_grid)[0, -1] for i in range(num_path)]


#compute call price
option1.otype = 1
payoff = [option1.payoff(s) for s in terminal_price]
call_value = np.exp(-r*option1.maturity)*np.average(payoff)
print('call price is ' + str(call_value))
print('exact price is 10.3009')
```

```python

#compute put price
option1.otype = -1
payoff = [option1.payoff(s) for s in terminal_price]
put_value = np.exp(-r*option1.maturity)*np.average(payoff)
print('put price is ' + str(put_value))
```
